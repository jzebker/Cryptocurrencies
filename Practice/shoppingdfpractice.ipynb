```python
import pandas as pd
file_path = 'shopping_data.csv'
df_shopping = pd.read_csv(file_path,encoding='ISO-8859-1')
df_shopping.head(5)
```

```python
# Columns
df_shopping.columns
```

```python
# Dtypes
df_shopping.dtypes
```

```python
# Drop null rows
df_shopping = df_shopping.dropna()
```

```python
# Drop CustomerID
df_shopping = df_shopping.drop(columns=["CustomerID"])
```

```python
# Check for dupes
df_shopping.duplicated().sum()
```

```python
# Change Card Member to int (1 or 0)
def str_to_bin(CardMember):
    if CardMember=='Yes':
        return 1
    else:
        return 0

df_shopping['Card Member']=df_shopping['Card Member'].apply(str_to_bin)
df_shopping.head()
```

```python
# Scale down Annual Income
df_shopping['Annual Income']=df_shopping['Annual Income']/1000
df_shopping.head()
```

```python
df_shopping=df_shopping.rename(columns={'Card Member':'CardMember','Annual Income':'AnnualIncome','Spending Score (1-100)':'SpendingScore'})
df_shopping
```

```python
# Saving cleaned data
file_path = "shopping_data_cleaned.csv"
df_shopping.to_csv(file_path, index=False)
```

```python
# Saving cleaned data
file_path = "shopping_data_cleaned.json"
df_shopping.to_json(file_path)
```
